```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# 1. Prompt (プロンプトテンプレート)の準備
prompt = ChatPromptTemplate.from_messages([
    ("system", "あなたはプロの要約作成者です。ユーザーが入力した文章を、50字以内の簡潔な要約にしてください。"),
    ("user", "{text_to_summarize}") # {text_to_summarize} の部分がユーザーの入力に置き換わります
])
```

```python
# 2. Model (AIモデル)の準備
model = ChatGoogleGenerativeAI(
      model="gemini-flash-latest",
    temperature=0 # 回答の一貫性を確保するため0に設定
)
```

```python
# 3. Output Parser (出力パーサー)の準備
output_parser = StrOutputParser()
```

```python
# 3つの部品をパイプ「|」でつなげてチェーンを構築
chain = prompt | model | output_parser

# チェーンを実行し、結果を表示する
sample_text = "LangChainはAIアプリケーション開発を容易にするフレームワークです。Gemini APIと組み合わせることで、高性能なチャットボットやデータ処理チェーンを迅速に構築できます。この統合により、開発者は複雑なAI処理を抽象化し、よりビジネスロジックに集中できるようになります。"
response = chain.invoke({"text_to_summarize": sample_text})
print(f"--- 元のテキスト ---\n{sample_text}\n")
print(f"--- 要約結果 ---\n{response}")
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
# 要件通り text-embedding-004 モデルを使用
embeddings_model = GoogleGenerativeAIEmbeddings(
    model="models/text-embedding-004"
)
```

```python
# 単一のテキストをベクトル化
vector = embeddings_model.embed_query("人工知能と機械学習")
print(f"--- ベクトル化された結果(先頭10次元) ---")
print(vector[:10])
print(f"\n--- ベクトルの総次元数---")
print(len(vector))
# text-embedding-004 は768次元のベクトルを生成することを確認してください。
```

```python
# 複数のテキストをリストで準備
documents = [
    "Pythonはデータサイエンスで広く使われます。",
    "LangChainはAI開発のためのフレームワークです。",
    "Google ColabはクラウドベースのJupyterノートブック環境です。"
]
# 複数のテキストを一度にベクトル化
doc_vectors = embeddings_model.embed_documents(documents)
print (f"--- ベクトル化された文書の数 ---")
print(len(doc_vectors))
print(f"\n--- 最初の文書のベクトルの次元数 ---")
print(len(doc_vectors[0]))
```

```python
print("--- ストリーミング出力開始---")
# streamメソッドで実行
stream_text = "ストリーミング実行は、大規模な回答を生成する際にユーザー体験を向上させることができます。回答が一度に表示されるのではなく、少しずつ画面に出力されるためです。"
for chunk in chain.stream({"text_to_summarize": stream_text}):
    print(chunk, end="", flush=True) # flush=True でリアルタイム表示
print("\n--- ストリーミング出力終了 ---")
```

```python
# 複数の入力をリストで準備
inputs = [
    {"text_to_summarize": "バッチ処理は、複数の独立したタスクをまとめて実行するのに適しています。これにより、API呼び出しのオーバーヘッドを削減できます。"},
    {"text_to_summarize": "クラウドコンピューティング技術は、現代のソフトウェア開発においてインフラストラクチャの柔軟性とスケーラビリティを提供します。"},
    {"text_to_summarize": "自然言語処理 (NLP) は、テキストデータの分析と理解に焦点を当てたAIの一分野です。"}
]
# batchメソッドで一度に処理
results = chain.batch(inputs)
print("--- バッチ処理の結果 ---")
for i, res in enumerate(results, 1):
    print(f"--- 結果 {i} ---")
    print(res)
```
